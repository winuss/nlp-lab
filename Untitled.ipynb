```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(-100, 100, 50).cumsum()

print(data)
#plt.rcParams['axes.unicode_minus'] = False

plt.plot(range(50), data, 'r')
plt.title('가격변동 추이')
plt.ylabel('가격')
plt.show()
```
